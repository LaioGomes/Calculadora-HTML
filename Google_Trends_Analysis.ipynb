```python
from pytrends.request import TrendReq
import yfinance as yf
import vectorbt as vbt
```

```python
pytrend= TrendReq()
```

```python
pytrend
```

```python
pytrend_BR = TrendReq(hl='pt-BR', tz= 180, geo = 'BR')
```

```python
pytrend_BR.trending_searches(pn='brazil')
```

```python
pytrend.top_charts(2021, hl= 'ptBR', geo = 'BR')
```

```python
keyword = ['Petrobras']
pytrend.build_payload(keyword, timeframe = '2022-01-01 2022-10-20', geo= '')
```

```python
petro = pytrend.interest_over_time()
petro
```

```python
petro.plot();
```
